```python
modelIDs=['kmeans','sc','hclust','birch']
modelNames=['K-Means','Spectral','Agglomerative','Birch']
rowNames=['Silhouette','Calinski-Harabasz','Davies-Bouldin']
nC=[3,4,5]
```

```python
for i in range(len(modelNames)):
  result=pd.DataFrame()
  result['Parameters']=rowNames
  setup(data=data,verbose=False)
  for j in nC:
    print('Number of Clusters =',j)
    model=create_model(modelIDs[i],num_clusters=j)
    r=pull()
    result['None_'+str(j)]=[round(float(r['Silhouette']),2),round(float(r['Calinski-Harabasz']),2),round(float(r['Davies-Bouldin']),2)]
  setup(data=data,normalize=True,normalize_method='zscore',verbose=False)
  for j in nC:
    print('Number of Clusters =',j)
    model=create_model(modelIDs[i],num_clusters=j)
    r=pull()
    result['Norm_'+str(j)]=[round(float(r['Silhouette']),2),round(float(r['Calinski-Harabasz']),2),round(float(r['Davies-Bouldin']),2)]
  setup(data=data,transformation=True,transformation_method='yeo-johnson',verbose=False)
  for j in nC:
    print('Number of Clusters =',j)
    model=create_model(modelIDs[i],num_clusters=j)
    r=pull()
    result['T_'+str(j)]=[round(float(r['Silhouette']),2),round(float(r['Calinski-Harabasz']),2),round(float(r['Davies-Bouldin']),2)]
  setup(data=data,pca=True,pca_method='linear',verbose=False)
  for j in nC:
    print('Number of Clusters =',j)
    model=create_model(modelIDs[i],num_clusters=j)
    r=pull()
    result['PCA_'+str(j)]=[round(float(r['Silhouette']),2),round(float(r['Calinski-Harabasz']),2),round(float(r['Davies-Bouldin']),2)]
  setup(data=data,transformation=True,transformation_method='yeo-johnson',normalize=True,normalize_method='zscore',verbose=False)
  for j in nC:
    print('Number of Clusters =',j)
    model=create_model(modelIDs[i],num_clusters=j)
    r=pull()
    result['T+N_'+str(j)]=[round(float(r['Silhouette']),2),round(float(r['Calinski-Harabasz']),2),round(float(r['Davies-Bouldin']),2)]
  setup(data=data,transformation=True,normalize=True,pca=True,normalize_method='zscore',transformation_method ='yeo-johnson',pca_method='linear',verbose=False)
  for j in nC:
    print('Number of Clusters =',j)
    model=create_model(modelIDs[i],num_clusters=j)
    r=pull()
    result['T+N+PCA_'+str(j)]=[round(float(r['Silhouette']),2),round(float(r['Calinski-Harabasz']),2),round(float(r['Davies-Bouldin']),2)]
  result.to_csv(modelNames[i]+'.csv',index=False)
```

```python
models=['ap','meanshift','dbscan','optics']
modelName=['Affinity','Mean Shift','Density-Based Spatial','OPTICS']
colNames=['No Data Processing','Using Normalization','Using Transform','Using PCA','Using T+N','Using T+N+PCA']
```

```python
for i in range(len(modelName)):
  result=pd.DataFrame()
  result['Parameters']=rowNames
  setup(data=data,verbose=False)
  model=create_model(models[i])
  r=pull()
  result[colNames[0]]=[round(float(r['Silhouette']),2),round(float(r['Calinski-Harabasz']),2),round(float(r['Davies-Bouldin']),2)]
  setup(data=data,normalize=True,normalize_method='zscore',verbose=False)
  model=create_model(models[i])
  r=pull()
  result[colNames[1]]=[round(float(r['Silhouette']),2),round(float(r['Calinski-Harabasz']),2),round(float(r['Davies-Bouldin']),2)]
  setup(data=data,transformation=True,transformation_method='yeo-johnson',verbose=False)
  model=create_model(models[i])
  r=pull()
  result[colNames[2]]=[round(float(r['Silhouette']),2),round(float(r['Calinski-Harabasz']),2),round(float(r['Davies-Bouldin']),2)]
  setup(data=data,pca=True,pca_method='linear',verbose=False)
  model=create_model(models[i])
  r=pull()
  result[colNames[3]]=[round(float(r['Silhouette']),2),round(float(r['Calinski-Harabasz']),2),round(float(r['Davies-Bouldin']),2)]
  setup(data=data,transformation=True,transformation_method='yeo-johnson',normalize=True,normalize_method='zscore',verbose=False)
  model=create_model(models[i])
  r=pull()
  result[colNames[4]]=[round(float(r['Silhouette']),2),round(float(r['Calinski-Harabasz']),2),round(float(r['Davies-Bouldin']),2)]
  setup(data=data,transformation=True,normalize=True,pca=True,normalize_method='zscore',transformation_method ='yeo-johnson',pca_method='linear',verbose=False)
  model=create_model(models[i])
  r=pull()
  result[colNames[5]]=[round(float(r['Silhouette']),2),round(float(r['Calinski-Harabasz']),2),round(float(r['Davies-Bouldin']),2)]
  result.to_csv(modelName[i]+'.csv',index=False)
```
